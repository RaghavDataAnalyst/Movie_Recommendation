```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
movies['details']=movies['details'].apply(lambda x:x.lower())
```

```python
vector = cv.fit_transform(movies['details']).toarray()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(x):
    y=[]
    for i in x.split():
     y.append(ps.stem(i))
     return ' '.join(y)

```

```python
ps.stem('Loving')
```

```python
movies['details']= movies['details'].apply(stem)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
simi =cosine_similarity(vector)
```

```python
simi    # It is showing similarity of 1 movie with other movies
```

```python
sorted(list (enumerate(simi[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def recommend(moviii):
   movie_index= movies[movies['title']==moviii].index[0]
   distance =simi[movie_index]
   movies_list =sorted(list (enumerate(distance)),reverse=True,key=lambda x:x[1])[1:6]
   for i in movies_list:
    print(movies.iloc[i[0]].title)
```

```python
recommend('Avatar')
```
